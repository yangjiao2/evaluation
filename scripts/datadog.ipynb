```python
def send_eval_metrics(eval_result):
    project = eval_result['project']
    eval_id = eval_result['eval-id']
    scores = eval_result['scores']
    # timestamp = eval_result['timestamp']      # let gauge handles this
    for metric, value in scores.items():
        # Sending as a GAUGE
        statsd.gauge(f"eval.{metric}", value, tags=[
            f"eval-id:{eval_id}",
            f"project:{project}"
        ])
        # Sending as a HISTOGRAM
        statsd.histogram(f"eval.{metric}_distribution", value, tags=[
            f"eval-id:{eval_id}",
            f"project:{project}"
        ])
```
