```python
import os



from ddtrace.llmobs import LLMObs
import ddtrace.llmobs.experimentation as dne


import os
import requests

def fetch_datadog_projects():
    """
    Fetches Datadog LLM-Observability projects using API and Application keys from environment variables.

    Returns:
        dict: The response JSON from Datadog API if successful, None otherwise.
    """
    url = "https://app.datadoghq.com/api/unstable/llm-obs/v1/projects"

    # Get API keys from environment variables
    api_key = os.getenv("DD_API_KEY")
    app_key = os.getenv("DD_APPLICATION_KEY")

    if not api_key or not app_key:
        raise ValueError("Missing API keys. Set 'DD_API_KEY' and 'DD_APPLICATION_KEY' in environment.")

    headers = {
        "DD-API-KEY": api_key,
        "DD-APPLICATION-KEY": app_key,
  "Accept": "application/json",
        "Content-Type": "application/json", 
    }

    response = requests.post(url, headers=headers, json={})  

    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error {response.status_code}: {response.text}")
        return None

result = fetch_datadog_projects()

```
