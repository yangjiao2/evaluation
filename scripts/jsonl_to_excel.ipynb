```python
import pandas as pd
import json

def jsonl_to_excel(jsonl_file_path, excel_file_path):
    # Initialize an empty list to store the JSON objects
    data = []
    
    # Open the JSONL file and read each line
    with open(jsonl_file_path, 'r') as file:
        for line in file:
            # Parse the JSON object and append it to the list
            data.append(json.loads(line))
    
    # Convert the list of JSON objects to a DataFrame
    df = pd.DataFrame(data)
    print (df)
    # Save the DataFrame to an Excel file
    df.to_excel(excel_file_path, index=False)

# Example usage
jsonl_file_path = 'data.jsonl'
excel_file_path = 'output.xlsx'
jsonl_to_excel(jsonl_file_path, excel_file_path)
```

```python

import concurrent.futures
import time

def task_function(task_id):
    print(f"Task {task_id} is starting...")
    time.sleep(2)  # Simulating a task taking 2 seconds
    print(f"Task {task_id} is completed.")
    return f"Result of task {task_id}"

# Function to create and manage the thread pool
def main():
    # Create a ThreadPoolExecutor with 4 worker threads
    with concurrent.futures.ThreadPoolExecutor(max_workers=4) as executor:
        # List of tasks to be run
        tasks = [1, 2, 3, 4, 5, 6, 7, 8]

        # Submit tasks to the executor
        future_to_task = {executor.submit(task_function, task): task for task in tasks}

        # Process completed tasks as they finish
        for future in concurrent.futures.as_completed(future_to_task):
            task = future_to_task[future]
            try:
                result = future.result()
            except Exception as exc:
                print(f"Task {task} generated an exception: {exc}")
            else:
                print(f"Task {task} returned: {result}")


main()

```
