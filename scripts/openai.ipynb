```python

import json

import requests

url = "https://prod.api.nvidia.com/oauth/api/v1/ssa/default/token"
headers = {"Content-Type": "application/json"}
d = {
    "client_id": "id",
    "client_secret": "secret",
    "scope": "openai-readwrite azureopenai-readwrite",
    "grant_type": "client_credentials"
}

response = requests.post(url, headers=headers, json=d)
api_key = json.loads(response.text)["access_token"]
from openai import AzureOpenAI

client = AzureOpenAI(
    azure_endpoint="https://prod.api.nvidia.com/llm/v1/azure/chat/completions",
    api_key=api_key,
    api_version="2023-07-01-preview",
)


```

```python
json_data = {
    "model": "o1-preview",  #self.model_config.llm_name,
    "messages": [{"role": "user", "content": "content"}],
    "frequency_penalty": 0,
    "logit_bias": None,
    "logprobs": False,
    "top_logprobs": None,
    # "max_tokens": 5000,
    "max_completion_tokens": 5000,
    "n": 1,
    "presence_penalty": 0,
    "response_format": {"type": "text"},
    "seed": 0,
    # "stop": None,
    "temperature": 1,
    "top_p": 1,
}
# json_data.update(extra_body)

model_config_base_url = "https://prod.api.nvidia.com/llm/v1/azure"
response = requests.post(
    model_config_base_url + "/chat/completions",
    headers={
        "Content-Type": "application/json",
        "Authorization": f"Bearer {api_key}",
    },
    json=json_data,
)
if response.status_code == 200:
    text = json.loads(response.content)["choices"][0]["message"][
        "content"
    ]
else:
    raise ValueError(response.status_code, response.content)
```
