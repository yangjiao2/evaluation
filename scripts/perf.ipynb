```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_grouped_bar_chart(
    data,
    id_vars,
    x,
    y_label,
    hue,
    col,
    palette="bright",
    figsize=(10, 5),
    rotation=25
):
    """
    Generalized grouped bar chart with faceting.

    Parameters:
        data: dict or pd.DataFrame - input data
        id_vars: list[str] - columns to keep fixed for melting
        x: str - column name for x axis (usually the melted variable column)
        y_label: str - label for y axis
        hue: str - column name for hue (color grouping)
        col: str - column name for facet columns
        palette: str or list - seaborn palette name or list of colors
        figsize: tuple - figure size per facet
        rotation: int - x-axis label rotation in degrees
    """
    # Convert to DataFrame if dict
    if isinstance(data, dict):
        df = pd.DataFrame(data)
    else:
        df = data.copy()

    # Melt the dataframe
    df_long = df.melt(id_vars=id_vars, var_name=x, value_name="Value")

    sns.set_theme(style="whitegrid")
    palette_colors = sns.color_palette(palette)

    g = sns.catplot(
        data=df_long,
        kind="bar",
        x=x,
        y="Value",
        hue=hue,
        col=col,
        palette=palette_colors,
        errorbar=None,  # changed here
        height=figsize[1],
        aspect=figsize[0]/figsize[1]
    )

    # Label bars
    for ax in g.axes.flat:
        for container in ax.containers:
            ax.bar_label(container, fmt="%.2f", label_type="edge", padding=2)

    # Titles and labels
    g.set_titles("{col_name} = {col_var}")  # fixed placeholder
    g.set_axis_labels("", y_label)

    # Rotate x labels
    for ax in g.axes.flat:
        ax.tick_params(axis="x", rotation=rotation)

    # Adjust legend outside to the right
    g._legend.set_bbox_to_anchor((1.05, 0.5))
    g._legend.set_loc('center left')
    g._legend.set_title(hue)

    plt.tight_layout()
    plt.show()


# Example usage with your data
data = {
    "Concurrency": [8, 8, 16, 16],
    "Group":       ["Bugnemo", "Optimized", "Bugnemo", "Optimized"],
    "Req Latency (s)":        [16.43,  9.10, 20.55, 13.93],
    "Req Throughput (req/s)": [0.22,   0.64, 0.23,  0.66],
    "TTFT (s)":               [7.54,   1.98, 10.82,  7.44],

    # "Token Throughput (tokens/s)": [117.67, 312.79, 133.29, 281.53],
}

plot_grouped_bar_chart(
    data=data,
    id_vars=["Concurrency", "Group"],
    x="Metric",
    y_label="Value (s or rate)",
    hue="Group",
    col="Concurrency",
    palette="bright",
    figsize=(8,5),
    rotation=25
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_performance_by_concurrency(data: dict, save_path=None):
    """
    Plot p99 TTFT, p99 Request Latency, and Request Throughput
    across different concurrency levels with even spacing.

    data: {
      "Concurrency": [...],
      "p99 TTFT (ms)": [...],
      "p99 Req Latency (ms)": [...],
      "Req Throughput (req/s)": [...],
    }
    save_path: optional path to save the plot as PNG.
    """
    df = pd.DataFrame(data).sort_values("Concurrency")
    positions = list(range(len(df)))
    width = 0.3

    fig, ax1 = plt.subplots(figsize=(10, 6))
    ax2 = ax1.twinx()

    # Bar plots with even spacing
    ax1.bar(
        [p - width for p in positions],
        df['p99 TTFT (ms)'],
        width,
        label='p99 TTFT (ms)',
        color='skyblue'
    )
    ax1.bar(
        [p + width for p in positions],
        df['p99 Req Latency (ms)'],
        width,
        label='p99 Request Latency (ms)',
        color='lightgreen'
    )

    # Throughput line plot
    ax2.plot(
        positions,
        df['Req Throughput (req/s)'],
        marker='o',
        label='Req Throughput (req/s)',
        color='darkorange'
    )

    # Axis formatting
    ax1.set_xticks(positions)
    ax1.set_xticklabels(df['Concurrency'])
    ax1.set_xlabel('Concurrency')
    ax1.set_ylabel('Latency (ms)')
    ax2.set_ylabel('Throughput (req/s)')

    # Legends and title
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    plt.title('Performance vs Concurrency (even spacing)')
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path)
    plt.show()


# --- Example usage ---
data = {
    "Concurrency": [1, 2, 4, 8, 16, 32, 64],
    "p99 TTFT (ms)": [892.74, 1243.13, 2527.24, 13398.12, 29573.70, 92735.75, 145965.73],
    "p99 Req Latency (ms)": [13672.74, 10340.62, 18607.70, 24467.60, 66337.37, 128678.44, 154621.45],
    "Req Throughput (req/s)": [0.15, 0.29, 0.35, 0.41, 0.32, 0.36, 0.44],
}

data = {
  "Concurrency": [1, 2, 4, 8, 16, 32, 64],
  "p99 TTFT (ms)": [863.42, 920.64, 1575.19, 3128.95, 50605.76, 21516.18, 74871.34],
  "p99 Req Latency (ms)": [9681.46, 19166.89, 21416.91, 16034.06, 72631.46, 43156.79, 90770.96],
   "Req Throughput (req/s)": [0.16, 0.27, 0.31, 0.78, 0.41, 1.04, 1.10]
}

# plot_performance_by_concurrency(data, save_path="performance_concurrency.png")
plot_performance_by_concurrency(data, save_path="performance_concurrency_op.png")
```
